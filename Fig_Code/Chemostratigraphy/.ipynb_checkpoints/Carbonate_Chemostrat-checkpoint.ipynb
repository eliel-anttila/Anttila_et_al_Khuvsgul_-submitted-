```python
Bow2022=pd.read_excel('Data/Global_Composites/Bow_22_Comp_Simple.xlsx')
Ed2022=pd.read_excel('Data/Global_Composites/Yang_et_al_2021_Comp_simple.xlsx')
```

```python
C=Bow2022['δ13Ccarb (‰)']
h=Bow2022['[Model C] age (Ma)']
```

```python
ECRanked=plt.figure(figsize=(10,20))#set figure dimensions
plt.scatter(C,h,s=10,color='#454545',zorder=2)
plt.axhline(525)

def carb_axes():
    plt.axvline(0, color='k')
    plt.axvline(4, color='k', ls='--',linewidth=.5)
    plt.axvline(-4, color='k',ls='--',linewidth=.5)
    
carb_axes()

plt.gca().set_ylim([515,545])
plt.gca().set_xlim([-12,12])
plt.gca().invert_yaxis()


plt.savefig('Output/Global_Composites/Bowyer_Camb.pdf')
plt.savefig('Output/Global_Composites/Bowyer_Camb.svg')
    
```
