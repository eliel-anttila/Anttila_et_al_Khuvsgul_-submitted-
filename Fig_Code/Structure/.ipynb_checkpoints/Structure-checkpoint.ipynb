```python
Plane2017=pd.read_csv('Plane_Data/plane_2017_1.csv')
Plane2018=pd.read_csv('Plane_Data/plane_2018.csv')
Plane2019=pd.read_csv('Plane_Data/plane_2019.csv')

Pl = pd.concat([Plane2017, Plane2018, Plane2019], axis=0)
```

```python
## sort for Khoridol Saridag Range and Darkhat Valley (between 50.5N and 51.25N) and only take bedding measurement type ()
```

```python
Pl_KSR=Pl.loc[(Pl[' latitude'] >= 50.5) & (Pl[' latitude']<= 51.25) & (Pl[' planeType']==' Bedding') & (Pl[' dip'] >= 20)]
Pl_KSR_Kheseen=Pl_KSR.loc[(Pl_KSR[' rockUnit'] ==' Lower Khesen')]


P2019_F=Plane2019.loc[(Plane2019[' planeType']==' Foliation')]
```

```python
#Isolate D1 structure bedding planes
Pl_KSR_D1=Pl.loc[(Pl[' latitude'] >= 50.58) & (Pl[' latitude']<= 50.72) & (Pl[' longitude'] >= 100.015) & (Pl[' longitude']<= 100.18)]
D1_S=Pl_KSR_D1[' strike'].mean()
D1_D=Pl_KSR_D1[' dip'].mean()

#Isolate D1 structure bedding planes
Pl_KSR_D1=Pl.loc[(Pl[' latitude'] >= 50.58) & (Pl[' latitude']<= 50.72) & (Pl[' longitude'] >= 100.015) & (Pl[' longitude']<= 100.18)]
D1_S=Pl_KSR_D1[' strike'].mean()
D1_D=Pl_KSR_D1[' dip'].mean()
```

```python
#create stereonet

import mplstereonet

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111, projection='stereonet')
cax = ax.density_contourf(Pl_KSR[' strike'], Pl_KSR[' dip'], measurement='poles',cmap='twilight') #plot all KSR+Darkhat_Data
ax.pole(Pl_KSR[' strike'], Pl_KSR[' dip'], color='k')
ax.pole(P2019_F[' strike'], P2019_F[' dip'], color='r')
ax.plane(D1_S,D1_D, color='goldenrod')
ax.grid(True)
fig.colorbar(cax)
```
